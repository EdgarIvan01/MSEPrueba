```python
df_features=df_c
array=df_features.values
x_class = array[ : , 0:3]
y_class = array[ : , 3]
X_trainlg, X_testlg, y_trainlg, y_testlg = train_test_split(x_class, y_class, test_size=0.2,random_state=0)
```

```python
df_features
```

```python
# Crear el modelo de Random Forest
rf_model = RandomForestRegressor(n_estimators=20, random_state=0)

# Entrenar el modelo
rf_model.fit(X_trainlg, y_trainlg)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Hacer predicciones con el conjunto de prueba
y_pred = rf_model.predict(X_testlg)

# Calcular el error cuadrático medio
mse = mean_squared_error(y_testlg, y_pred)
print(f'Error cuadrático medio (MSE): {mse:.2f}')

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_testlg, y_pred)
print(f'Coeficiente de determinación (R^2): {r2:.2f}')
```
